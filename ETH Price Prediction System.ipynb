```python
# Description : This program predicts the price of ETH
```

```python
#Import the dependencies
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Load the data
from google.colab import files
files.upload()
```

```python
#Get and show the data
#Store the data
df = pd.read_csv('ETH.csv')
#Show the data
df
```

```python
#Create a variable for predicting 'n' days
projection = 14
#Create a new coloumn called prediction
df['Prediction'] = df[['Close']].shift(-projection)
#Show the data set
df
```

```python
#Create the independent data set (x)
X = np.array(df[['Close']])
#Remove the last 14 rows
X = X[:-projection]
print(X)
```

```python
#Create the dependent data set (y)
y = df['Prediction'].values
y = y[:-projection]
print(y)
```

```python
#Split the data into 85% tarining and 15% testing data sets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = .15)
```

```python
#Create and train the model
linReg = LinearRegression()
#Train the model
linReg.fit(x_train, y_train)
```

```python
#Test the model using score
linReg_confidence = linReg.score(x_test, y_test)
print('Linear regression Confidence:', linReg_confidence)
```

```python
#Create a variable called x_projection and set it equal to the last 14 rows of data from the original data set
x_projection = np.array(df[['Close']])[-projection:]
print(x_projection)
```

```python
#Print the linear regression models prediction for the next 14 days
linReg_prediction = linReg.predict(x_projection)
print(linReg_prediction)
```
